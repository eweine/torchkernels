```python
from torchkernels.feature_maps import LaplacianORF, LaplacianRFF
from utils import create_x1_x2_list, create_Kmat_exact, plot_kernel_approximation
```

```python
input_dim = 32
num_features = int(3e4)
length_scale = 1.
```

```python
x1_list, x2_list, x1_x2_norm = create_x1_x2_list(input_dim=input_dim, N=20)
K_exact_list = create_Kmat_exact(kernels=['Laplacian'], x1_list=x1_list, x2_list=x2_list, length_scale=length_scale)['Laplacian']
```

```python
#ORF
K_approx_list_ORF = []
feature_map = LaplacianORF(input_dim=input_dim, num_features=num_features, length_scale=length_scale)
for x1,x2 in zip(x1_list, x2_list):
    Phi_x1 = feature_map(x1)
    Phi_x2 = feature_map(x2)
    K_approx_list_ORF.append((Phi_x1@Phi_x2.T)[0].item())
plot_kernel_approximation(x1_x2_norm, K_exact_list, K_approx_list_ORF, kernel_str='Laplacian - ORF')
```

```python
#RFF
K_approx_list_ORF = []
feature_map = LaplacianRFF(input_dim=input_dim, num_features=num_features, length_scale=length_scale)
for x1,x2 in zip(x1_list, x2_list):
    Phi_x1 = feature_map(x1)
    Phi_x2 = feature_map(x2)
    K_approx_list_ORF.append((Phi_x1@Phi_x2.T)[0].item())
plot_kernel_approximation(x1_x2_norm, K_exact_list, K_approx_list_ORF, kernel_str='Laplacian - RFF')
```
